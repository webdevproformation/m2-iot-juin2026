dessiner des graphiques

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dessiner des schemas 

# trait plein dans des axes x, y

tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_y = np.random.randint(0,100, ( 100, 1   ))

plt.plot(tab1_x, tab1_y , ls="-." )
# lw line width
# ls line style
# color
# alpha = opacité des dessins
plt.scatter(tab1_x, tab1_y , lw=2 , color="red" , alpha=0.4)
```

possibilité de faire dessiner PLUSIEURS graphiques par l'outil : `subplot()`

```python
tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_ytemperature = np.random.randint(0,100, ( 100, 1   ))
tab1_ypression = tab1_x ** 2 + 3

# créer des sous graphiques
plt.figure() # créer l'image en intégralité

# premier sous graphique 
plt.subplot(2,1,1) # grille qui contient 2 lignes 
                   # sur 1 une seule colonne
                   # je suis actuellement en train de session le 1er graphique
plt.plot(tab1_x, tab1_ytemperature , color="red" , label="température" )

plt.xlabel("temps")
plt.ylabel("temperature")
plt.title("graph 1")
plt.legend()

plt.subplot(2,1,2)
plt.title("graph 2")
plt.plot(tab1_x, tab1_ypression , color="green" , label="pression" )
plt.xlabel("temps")
plt.ylabel("pression")
plt.tight_layout() # espacement entre les graphiques empilés verticalement
plt.legend()
```

```python
# POO
np.random.seed(0)
tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_ytemperature = np.random.randint(0,100, ( 100, 1   ))

fig, ax = plt.subplots(2,1, sharex=True) # Attention au s 

ax[0].plot(tab1_x , tab1_ytemperature , c="g")
ax[1].plot(tab1_x , tab1_ytemperature * -1 , c="r")
```

```python
datasets = { f"releve {i}" : np.random.randn(100) for i in range(0,5) }

def generate_graph( datasets ):
    n = len(datasets)
    plt.figure(figsize=(12,8))
    for i , v in zip( range( 1, n + 1), datasets.keys() ):
        plt.subplot(n , 1 , i )
        plt.plot( np.linspace(0,100, 100) , datasets[v] )
        plt.title(v)
    plt.tight_layout()
    
generate_graph( datasets )
```

```python
import numpy as np
f = lambda x, y : np.sin(x) + np.cos(x + y) # ce que l'on veut représenter en 3d

X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)

X, Y = np.meshgrid(X,Y)

Z = f(X, Y)

ax = plt.axes(projection="3d")
ax.plot_surface(X, Y , Z , cmap="plasma")
```

```python
f = lambda x, y : np.sin(x) + np.cos(x + y)

X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)

X, Y = np.meshgrid(X,Y)

Z = f(X, Y)

plt.contour( X,Y , Z , 20 , cmap="RdGy")
plt.colorbar()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris() # le dataset

x = iris.data
y = iris.target # 0 1 2

def generateFigure(datasets , y):
    n = datasets.shape[1]
    plt.figure(figsize=(12,10))
    for i in range(n):
        plt.subplot( n//2 , n//2 , i + 1 )
        plt.scatter( datasets[: , 0] , datasets[:,i] , c=y , alpha=0.5  )
        plt.xlabel("longueur du sépale")
        plt.ylabel(i)
    plt.tight_layout()
generateFigure(x , y)
```

```python
import math
import random 

import matplotlib.pyplot as plt 

nb_point= 25
x =      [ random.triangular()                 for i in range (nb_point)  ]
y =      [ random.gauss(0.5, 0.25)             for i in range (nb_point)  ]
color =  [ random.randint(1,4)                 for i in range (nb_point)  ]
taille = [ math.pi *random.randint(5,15) ** 2  for i in range (nb_point)  ]

# dessiner notre graphique

plt.figure(figsize=(12,8))
plt.scatter( x , y , c = color , alpha=0.5 , s=taille)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
import numpy as np
from matplotlib import pyplot as plt

num_set = [{'Language':75, 'Science':88, 'Math':96},
           {'Language':71, 'Science':95, 'Math':92},
           {'Language':75, 'Science':90, 'Math':89}]

lan_guage    = [['Language','Science','Math'], 
               ['Science','Math','Language'], 
               ['Math','Language','Science']] 
colors = ["r","g","b"]
names = sorted(num_set[0].keys())
values = np.array([[data[name] for name in order] for data,order in zip(num_set, lan_guage)])
lefts = np.insert(np.cumsum(values, axis=1),0,0, axis=1)[:, :-1]
orders = np.array(lan_guage)
bottoms = np.arange(len(lan_guage))

for name, color in zip(names, colors):
    idx = np.where(orders == name)
    value = values[idx]
    left = lefts[idx]
    plt.bar(height=0.8, x=value,width=value, bottom=bottoms, color=color, orientation="horizontal", label=name)
	
plt.yticks(bottoms+0.4, ["Student-%d" % (t+1) for t in bottoms])
plt.legend(loc="best", bbox_to_anchor=(1.0, 1.00))
plt.subplots_adjust(right=0.75)
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

```
