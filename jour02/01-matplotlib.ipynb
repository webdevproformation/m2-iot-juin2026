dessiner des graphiques

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dessiner des schemas 

# trait plein dans des axes x, y

tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_y = np.random.randint(0,100, ( 100, 1   ))

plt.plot(tab1_x, tab1_y , ls="-." )
# lw line width
# ls line style
# color
# alpha = opacité des dessins
plt.scatter(tab1_x, tab1_y , lw=2 , color="red" , alpha=0.4)
```

possibilité de faire dessiner PLUSIEURS graphiques par l'outil : `subplot()`

```python
tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_ytemperature = np.random.randint(0,100, ( 100, 1   ))
tab1_ypression = tab1_x ** 2 + 3

# créer des sous graphiques
plt.figure() # créer l'image en intégralité

# premier sous graphique 
plt.subplot(2,1,1) # grille qui contient 2 lignes 
                   # sur 1 une seule colonne
                   # je suis actuellement en train de session le 1er graphique
plt.plot(tab1_x, tab1_ytemperature , color="red" , label="température" )

plt.xlabel("temps")
plt.ylabel("temperature")
plt.title("graph 1")
plt.legend()

plt.subplot(2,1,2)
plt.title("graph 2")
plt.plot(tab1_x, tab1_ypression , color="green" , label="pression" )
plt.xlabel("temps")
plt.ylabel("pression")
plt.tight_layout() # espacement entre les graphiques empilés verticalement
plt.legend()
```

Utiliser Matplotlib version POO `subplots()`

```python
# procédurale
np.random.seed(0)
tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_ytemperature = np.random.randint(0,100, ( 100, 1   ))

plt.plot(tab1_x, tab1_ytemperature)
```

```python
# POO
np.random.seed(0)
tab1_x = np.arange(0,100 , 1) # vection qui contient des chiffres de 0 à 100 avec un pas de 1
tab1_ytemperature = np.random.randint(0,100, ( 100, 1   ))

fig, ax = plt.subplots(2,1, sharex=True) # Attention au s 

ax[0].plot(tab1_x , tab1_ytemperature , c="g")
ax[1].plot(tab1_x , tab1_ytemperature * -1 , c="r")
```

exo

```python
datasets = { f"releve {i}" : np.random.randn(100) for i in range(0,5) }

def generate_graph( datasets ):
    n = len(datasets)
    plt.figure(figsize=(12,8))
    for i , v in zip( range( 1, n + 1), datasets.keys() ):
        plt.subplot(n , 1 , i )
        plt.plot( np.linspace(0,100, 100) , datasets[v] )
        plt.title(v)
    plt.tight_layout()
    
generate_graph( datasets )
```
